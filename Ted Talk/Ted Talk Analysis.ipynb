# Q6) Plot the Number of Ted Talk took place Each Year

```python
pd.to_datetime(df["film_date"], unit='s')
df["film_year"] = pd.to_datetime(df["film_date"], unit='s').dt.year
df["film_month"] = pd.to_datetime(df["film_date"], unit='s').dt.month
```

```python
df.head(1)
```

```python
df["film_year"].unique()
```

#### we have following Year in dataset
2006, 2004, 2005, 2003, 2007, 2002, 2001, 2008, 1984, 1990, 1998,
       2009, 1991, 2010, 1983, 1972, 2011, 2012, 1994, 2013, 2014, 2015,
       2016, 2017

```python

a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
sns_plot = sns.countplot(x=df["film_year"], data=df,ax=ax,)
plt.title("Ted Talk Count Each Year")
```

### Conclusion : as you can see TED talk count is increasing every Year
